We will explore how heat is distributed in the lithosphere, and how this heat influences lithospheric strength. You may remember that the primary sources of heat in Earths interior are from the decay of radioactive elements, and primordial heat remaining from Earths formation. Heat is also produced from frictional sliding on faults and chemical changes, but this is a very small component (< 1%) that we will not focus on today.

Heat is primarily transferred within Earths interior by convection and conduction. Convection requires the physical movement of material to facilitate cooling, conduction occurs in static (immobile) materials through the diffusive transfer of energy along a temperature gradient. Heat conduction can be mathematically modeled following a simple relationship first described by the French mathematician Jean-Baptiste Joseph Fourier.
